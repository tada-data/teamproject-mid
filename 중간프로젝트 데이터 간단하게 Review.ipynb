```python
OTT = pd.read_excel("data/가장많이이용하는OTT서비스(20-21).xlsx", header=[0,1], index_col=[0,1])
OTT
```

```python
OTT.columns =[["2020","2020","2020","2020","2020","2020","2020","2020","2020","2021","2021","2021","2021","2021","2021","2021","2021","2021","2021"], ["사례수","국내TV", "해외TV", "실시간TV", "영화", "OTT자체제작", "개인방송(예능/오락)", "음악", "기타","사례수", "영화", "개인방송(예능/오락)", "드라마","교양/다큐","교육/학습","애니메이션","스포츠","뉴스","기타"]]
```

```python
OTT
```

```python
OTT["시청증가율"] = OTT["2021"]["사례수"]/OTT["2020"]["사례수"] *100
```

```python
OTT
# 2020년에 비해 2021년의 OTT시청 사례수는 30-39세를 제외하고 모든 연령대가 증가하였다. 
```

```python
OTT_movie = pd.DataFrame([OTT["2020"]["영화"], OTT["2021"]["영화"]], index=["2020","2021"]).T
```

```python
OTT_movie["OTT영화시청증가율"]=OTT_movie["2021"]/OTT_movie["2020"]*100
```

```python
OTT_movie
```

```python
OTT_개인방송 = pd.DataFrame([OTT["2020"]["개인방송(예능/오락)"], OTT["2021"]["개인방송(예능/오락)"]], index=["2020","2021"]).T
```

```python
OTT_개인방송["OTT개인방송시청증가율"]=OTT_개인방송["2021"]/OTT_개인방송["2020"]*100
OTT_개인방송
```

```python
OTT.rename(columns = {"시청증가율":"OTT전체시청증가율"}, inplace=True)
```

```python
# OTT시청의 증감률을 확인하고 우리가 알아보고자 하는 OTT에서 영화를 보는 시청 증감률이 차이를 비교하기 위해 dataframe을 합쳐준다
# 추가로 2020년과 2021년의 공통 컬럼인 개인방송부분도 추가해서 비교하겠다.
```

```python
change = pd.concat([OTT_movie["OTT영화시청증가율"], OTT["OTT전체시청증가율"], OTT_개인방송["OTT개인방송시청증가율"]], axis=1)
```

```python
# OTT전체시청증감률 대비 OTT영화시청증감률을 비교한다. 

moviechange = change["OTT영화시청증가율"]/change["OTT전체시청증가율"] 
round(np.mean(moviechange),2)
# 전체적으로 OTT시청하는 사람들의 사례수가 증가한 추세인데, OTT에서 영화를 보는 사람들은 OTT전체 시청률대비에도 평균적으로 1.73배 높다. 

personal = change["OTT개인방송시청증가율"]/change["OTT전체시청증가율"] 
round(np.mean(personal),2)
# 한편 공통 컬럼으로 비교하고자 했던 개인방송을 시청하는 사람들은 OTT전체 시청률 대비 0.93배이다.
```
